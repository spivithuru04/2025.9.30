```python
column_titles = data_set.columns

numbers = [i for i in column_titles if data_set[i].dtype != 'object']
objects = [i for i in column_titles if data_set[i].dtype == 'object']
```

```python
numbers_data = data_set.drop(columns = numbers, axis = 1)
```

```python
new_data = pd.get_dummies(data_set, columns = objects, dtype = int)
```

```python
from sklearn.preprocessing import StandardScaler

for i in numbers:
    scaler = StandardScaler()
    new_data[i]= scaler.fit_transform(new_data[[i]])
```
